```python
from time import sleep
from kafka import KafkaConsumer
from confluent_avro import AvroKeyValueSerde, SchemaRegistry
```

```python
print("Connecting to consumer ...")
consumer = KafkaConsumer(
    'sails.sails.orders',
     bootstrap_servers=['localhost:9094'],
     auto_offset_reset='earliest',
     #enable_auto_commit=False,
     #group_id='test-group',
     )
registry_client = SchemaRegistry("http://0.0.0.0:8081",headers={"Content-Type": "application/vnd.schemaregistry.v1+json"})
avroSerde = AvroKeyValueSerde(registry_client, 'sails.sails.orders')
```

```python
for message in consumer:
    print(message)
```

```python
for msg in consumer:
    v = avroSerde.value.deserialize(msg.value)
    k = avroSerde.key.deserialize(msg.key)
    print(msg.offset, msg.partition, k, v)
```

```python
consumer.close()
```
